```python
import matplotlib.pyplot as plt;
```

```python
y = [[]];
x = [1,2,3,4]; # number of nodes
y.append([16,22,27,27]);
y.append([10,18,20,18]);
```

```python
fig, ax = plt.subplots(1);
ax.set_xticks(x); 
ax.set_yticks([0, 5, 10, 15, 20, 25, 30]);
ax.set_ylim([0,35])
ax.plot(x, y[1], '--ro', label='without bypassing');
ax.plot(x, y[2], '--bo', label='with bypassing');
ax.legend();
ax.set_xlabel('Testcase number'); 
ax.set_ylabel('Number of cycles'); 

```
