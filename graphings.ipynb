```python
import matplotlib.pyplot as plt;
```

```python

y = [16,22,27,27]; # number of nodes
x = [1,2,3,4];
plt.bar(x, y, width=0.4)
plt.xlabel("INSTRUCTION SET NUMBER")
plt.ylabel("CYCLE NUMBER")
plt.title("BAR PLOTS for PART-A")
plt.yticks(y)
plt.xticks(x)
plt.show()
```

```python
y = [10,18,20,18]; # number of nodes
x = [1,2,3,4]
plt.bar(x, y ,width=0.4,color = "red")
plt.xlabel("INSTRUCTION SET NUMBER")
plt.ylabel("CYCLE NUMBER")
plt.title("BAR PLOTS for PART-B")
plt.yticks(y)
plt.xticks(x)
plt.show()
```
